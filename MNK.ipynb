## Аппроксимция методом МНК

 Считаем данные,содержащиеся в таблице excel, с помощью бибилиотеки pandas

```python
import pandas as pd
url='https://github.com/sksucha/MSU2023/blob/main/nomer1.xlsx?raw=true'
data=pd.read_excel(url)

x=data['x']
y=data['y']
N=len(x)

data.head(35)
```

Найдем искомую функицю f=kx+b

```python
import numpy as np
mx = x.sum()/N
my = y.sum()/N

k = sum((x - mx)*(y - my))/sum((x-mx)**2)
b = my - k*mx
f = k*x+b
```

Воспользуемся бибилиотекой matplotlib для построения графика функции найденной аппроксимирующей функции f и обозначения начальных данных

```python
import matplotlib.pyplot as plt
plt.scatter(x, y, c='red')
plt.title('Аппроксимация методом МНК')
plt.xlim([x.min(),x.max()])
plt.ylim([y.min(),y.max()])
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,f, c='green')

plt.show()
```

Использование метода наименьших квадратов позволяет описать эксперементальные данные, имеющие некоторые разброс в значениях в силу погрешностей приборов или других неустранимых погрешностей. Наглядно представить результат работы алгоритмов Python помогает популярная бибилиотека Pyton для визуалицзации данных Matplotlib.

# Интерполяционный многочлен Лагранжа

Считаем данные,содержащиеся в таблице excel, с помощью бибилиотеки pandas

```python
import pandas as pd
url='https://github.com/sksucha/MSU2023/blob/main/Nomer2.xlsx?raw=true'
data=pd.read_excel(url)

data.head(35)
```

Для удобства работы с данными, удалим все строки с пропущенными значениями и сбросим исходные индексы в pandas data (Указание True запрещает pandas сохранять исходный индекс в виде столбца в data)

```python
data = data.dropna()
data = data.reset_index(drop=True)
```

```python
# Присваиваем новым переменным значения по столбцам нашего файла
x = data['Pres']
y = data['Salinity']
```

Построим искомый многочлен Лагранжа

```python
import numpy as np
x1 = np.linspace(x.min(), x.max(), 60)
N = len(x)
N1 = len(x1)
y1= np.zeros(N1)

for n in range(N1):
    s = 0
    for i in range(N):
        p = 1
        for j in range(N):
            if i != j:
                p = (x1[n] - x[j]) / (x[i] - x[j])*p
        s = s+ y[i] * p
    y1[n] = s
```

Воспользуемся бибилиотекой matplotlib для построения графика построенного полинома

```python
import matplotlib.pyplot as plt
plt.scatter(x, y)
plt.plot(x1, y1)
plt.title('Многочлен Лагранжа')
plt.xlabel('Pres')
plt.ylabel('Salinity')
plt.xlim(0)
plt.ylim(0, y.max())
plt.show()
```

Многочлен Лагранжа используется как многочлен минимальнй степени, принимающий заданные значения в заданном наборе точек, то есть решающий задачу интерполяции. С помощью бибилиотеки Matplotlib мы вновь построили график функций, решающей нашу опредлённую задачу.
