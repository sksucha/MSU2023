```python
import pandas as pd
url='https://github.com/sksucha/MSU2023/blob/main/nomer1.xlsx?raw=true'
data=pd.read_excel(url)

data.head(35)
```

```python
x=data['x']
y=data['y']
N=len(x)
```

```python
import numpy as np
mx = x.sum()/N
my = y.sum()/N

k = sum((x - mx)*(y - my))/sum((x-mx)**2)
b = my - k*mx
f = k*x+b
```

```python
import matplotlib.pyplot as plt
plt.scatter(x, y, c='red')
plt.xlim([x.min(),x.max()])
plt.ylim([y.min(),y.max()])
plt.plot(x,f, c='blue')

plt.show()
```
