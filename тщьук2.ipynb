```python
import numpy as np
x=data['Salinity']
y=data['Pres']
N=len(x)
```

```python
y1 = np.linspace(y.min(), y.max(), 60)
q = len(y)
m = len(y1)
x1 = np.zeros(m)

for n in range(m):
    s = 0
    for i in range(q):
        p = 1
        for j in range(q):
            if i != j:
                p *= (y1[n] - y[j]) / (y[i] - y[j])
        s += x1[i] * p
    x1[n] = s
```

```python
import matplotlib.pyplot as plt
plt.scatter(y, x)
plt.plot(y1, x1)
plt.title('Многочлен Лагранжа')
plt.xlabel('Pres')
plt.ylabel('Salinity')
plt.xlim(0)
plt.ylim(0, x.max())
plt.show()
```
